```python
## 1. Divide dataset into independent variables and dependent variables
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
## 2. Divide the dataset into test dataset and train dateset.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=36)
```

```python
## 3. Frature scaling -- Standardization for numerical features and OneHotEncoding for categorical features

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
print(scaler.fit(X_train))
```

```python
scaler.transform(X_train)
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_train
```

```python
X_test=scaler.transform(X_test)
```

```python
X_test
```

```python
## 4. Model Trianing
from sklearn.linear_model import LinearRegression
linear_regression=LinearRegression()
```

```python
## 5. Model Fit
linear_regression.fit(X_train,y_train)
```

```python
## 6. Coefficients and Intercept
linear_regression.coef_
```

```python
linear_regression.rank_
```

```python
linear_regression.n_features_in_
```

```python
linear_regression.singular_
```

```python
linear_regression.intercept_
```

```python
## 7. pridiction
y_pred=linear_regression.predict(X_train)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
import numpy as np
mse=mean_squared_error(y_test,y_pred)
print(mse)
mas=mean_absolute_error(y_test,y_pred)
print(mas)
rmse=np.sqrt(mse)
print(rmse)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
import numpy as np
mse=mean_squared_error(y_test,y_pred)
print(mse)
mae=mean_absolute_error(y_test,y_pred)
print(mae)
print(np.sqrt(mse))
```
